```python
X = df.drop("loan_status", axis=1)
y = df["loan_status"]
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object"]).columns

print("Numerical columns:", num_cols)
print("Categorical columns:", cat_cols)

```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Numeric pipeline
num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

# Categorical pipeline
cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_pipeline, num_cols),
        ("cat", cat_pipeline, cat_cols)
    ]
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
X_train_p = preprocessor.fit_transform(X_train)
X_test_p = preprocessor.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_p, y_train)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    random_state=42
)

xgb.fit(X_train_res, y_train_res)
```

```python
from sklearn.metrics import classification_report, roc_auc_score

y_pred = xgb.predict(X_test_p)
y_prob = xgb.predict_proba(X_test_p)[:, 1]

print(classification_report(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
X_train_dense = X_train_res
X_test_dense = X_test_p
```

```python
import shap

masker = shap.maskers.Independent(X_train_dense)
```

```python
explainer = shap.Explainer(
    xgb.predict_proba,
    masker,
    output_names=["Safe", "Risky"]
)
```

```python
shap_values = explainer(X_test_dense)
```

```python
from sklearn.pipeline import Pipeline
from xgboost import XGBClassifier

final_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", XGBClassifier(
        n_estimators=300,
        max_depth=6,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric="logloss",
        random_state=42
    ))
])
```

```python
final_model.fit(X_train, y_train)
```

```python
import os

os.makedirs("model", exist_ok=True)
```

```python
import joblib

joblib.dump(final_model, "model/xgb_model.pkl")
```

```python
feature_names = X_train.columns.tolist()
```

```python
def predict_fn(X):
    X_df = pd.DataFrame(X, columns=feature_names)
    return final_model.predict_proba(X_df)[:, 1]
```

```python
X_background = X_train.sample(100, random_state=42)
X_test_sample = X_test.sample(200, random_state=42)
```

```python
import shap
import matplotlib.pyplot as plt

explainer = shap.KernelExplainer(
    predict_fn,
    X_background
)
```

```python
shap_values = explainer.shap_values(X_test_sample)
```

```python
# Summary plot
plt.figure()
shap.summary_plot(shap_values, X_test_sample, show=False)
plt.tight_layout()
plt.savefig("shap_summary.png")
plt.close()

# Bar plot
plt.figure()
shap.summary_plot(
    shap_values,
    X_test_sample,
    plot_type="bar",
    show=False
)
plt.tight_layout()
plt.savefig("shap_feature_importance.png")
plt.close()
```
